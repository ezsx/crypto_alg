Проведения хэш-преобразования по алгоритму Стрибог (ГОСТ 34.11-2012) для произвольного набора длины 500 бит.

**Описание задачи:**

Мы реализуем хэш-преобразование для набора данных длиной 500 бит с использованием алгоритма Стрибог (ГОСТ 34.11-2012). Алгоритм Стрибог является одним из стандартов хэширования, принятым в России, и используется для обеспечения целостности данных.

**Пояснение решения:**

1. **Конвертация текста в бинарную строку:**
   - Функция `text_bin` принимает строку текста и возвращает ее бинарное представление. Она использует генератор списков для преобразования каждого символа в его ASCII код и затем в бинарное представление, заполняя нулями до восьми бит.
   - Функция `text_to_bin` выполняет ту же функцию, но использует метод `bytearray` для конвертации текста в бинарный массив.
   
2. **Конвертация бинарной строки в шестнадцатеричное представление и наоборот:**
   - Функция `bin_hex` принимает бинарную строку и возвращает ее шестнадцатеричное представление. Она разбивает бинарную строку на байты по восемь бит и преобразует каждый байт в шестнадцатеричное представление, заполняя его нулями при необходимости.
   - Функция `hex_bin` принимает шестнадцатеричное представление и возвращает его бинарное представление. Она разбивает шестнадцатеричное представление на байты по два символа и преобразует каждый байт в бинарное представление, заполняя его нулями до восьми бит.

3. **Реверсирование текста:**
   - Функция `reverse` принимает строку и возвращает ее в обратном порядке.

4. **Основные преобразования (XOR, S, P, L, и др.):**
   - Функция `x_change` выполняет побитовое исключающее ИЛИ над двумя строками заданной длины.
   - Функция `S` применяет нелинейное биективное преобразование к строке.
   - Функция `P` выполняет перестановку байтов в строке.
   - Функция `L` разбивает строку на блоки по 64 бита.
   - Функция `l` применяет линейное преобразование к блокам.
   - Функция `E` выполняет основную операцию преобразования.
   - Функция `key_schedule` выполняет расписание ключей.
   - Функция `g` выполняет функцию сжатия.
   - Функция `stribog_both` является главной функцией, которая применяет алгоритм Стрибога к заданному сообщению.

5. **Вспомогательные функции:**
   - Функция `text_2_approp` конвертирует текст в соответствующее представление.
   - Функция `text_to_hex` выполняет преобразование текста в его шестнадцатеричное представление.

### Алгоритм хэширования Стрибог (ГОСТ Р 34.11-2012):

**Пояснение:**

Алгоритм хэширования Стрибог (ГОСТ Р 34.11-2012) является стандартом хэширования, разработанным в России. Он работает с блоками данных размером 512 бит и выдает хэш-значение размером 256 бит. Алгоритм включает в себя несколько раундов, в которых применяются нелинейные преобразования к блоку данных.

**Процесс работы:**

1. **Инициализация:** Начальное значение (инициализационный вектор) загружается в состояние алгоритма.

2. **Добавление данных:** Входной текст разбивается на блоки по 512 бит. Каждый блок данных добавляется к текущему состоянию алгоритма.

3. **Сжатие данных:** Применяются серия нелинейных и линейных преобразований к состоянию алгоритма для сжатия данных и обеспечения стойкости хэша.

4. **Формирование хэша:** По завершении всех раундов обработки всех блоков данных, полученное состояние алгоритма является хэш-значением итогового хэша.

### Простая версия алгоритма MD5:

**Пояснение:**

MD5 (Message Digest Algorithm 5) - это широко используемый алгоритм хэширования, который создает хэш-значение фиксированной длины (128 бит). Он работает с блоками данных размером 512 бит и проходит через серию раундов для смешивания и преобразования данных.

**Процесс работы:**

1. **Инициализация:** Начальное значение (инициализационный вектор) загружается в состояние алгоритма.

2. **Предварительная обработка данных:** Входной текст разбивается на блоки по 512 бит. Каждый блок данных проходит через серию операций, включая добавление бита заполнения, добавление длины и другие манипуляции, чтобы привести его к размеру, необходимому для обработки алгоритмом.

3. **Сжатие данных:** Происходит серия раундов, в которых данные смешиваются и преобразуются с использованием нелинейных функций и таблиц констант.

4. **Формирование хэша:** После обработки всех блоков данных полученное состояние алгоритма является итоговым хэш-значением.

### Псевдокод алгоритма хэширования Стрибог (ГОСТ Р 34.11-2012):

```
1. Инициализация состояния алгоритма (IV)
2. Для каждого блока данных:
     2.1. Добавить блок данных к текущему состоянию алгоритма
     2.2. Применить нелинейные и линейные преобразования к состоянию
3. Получить хэш-значение из состояния алгоритма
```
